```python
orders_df['order_date'] = pd.to_datetime(orders_df['order_date'])
```

```python
order_value = order_items_df.groupby('order_id')['total_price'].sum().reset_index()
```

```python
order_value.rename(columns={'total_price':'order_value'},inplace = True)
```

```python
orders_merged = pd.merge(orders_df, order_value, on='order_id')
```

```python
reference_date = orders_merged['order_date'].max() + pd.Timedelta(days=1)
```

```python
rfm = orders_merged.groupby('customer_id').agg({'order_date': lambda x: 
(reference_date - x.max()).days, 'order_id': 'count','order_value': 'sum'}).reset_index()
```

```python
rfm.columns = ['customer_id', 'recency', 'frequency', 'monetary']
```

```python
rfm
```

```python
conn = sqlite3.connect("capstone_1")
```

```python
rfm.to_sql("rfm", con=engine, if_exists="replace", index=False)
```
